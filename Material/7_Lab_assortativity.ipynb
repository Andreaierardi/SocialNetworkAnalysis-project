```python
import networkx as nx
from statsmodels.distributions.empirical_distribution import ECDF
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example 8.3 Zafarani's book. Assortativity for nominal attributes. Disassortative network.
g = nx.Graph()
g.add_nodes_from([1,2],color='green')
g.add_nodes_from([3,4],color='blue')
g.add_edges_from([(2,3),(2,4),(1,3),(1,4)])
posizione = nx.spring_layout(g)
nx.draw_networkx(g,pos=posizione)
```

```python
print(nx.attribute_assortativity_coefficient(g,'color')) #normalized assortativity or assortativity coefficient
```

```python
# Example assortativity for nominal attributes. Assortative network.
g_ass = nx.Graph()
g_ass.add_nodes_from([1,2],color='green')
g_ass.add_nodes_from([3,4],color='blue')
g_ass.add_edges_from([(1,2),(3,4)])
posizione = nx.spring_layout(g_ass)
nx.draw_networkx(g_ass,pos=posizione)
print(nx.attribute_assortativity_coefficient(g_ass,'color')) #normalized assortativity or assortativity coefficient
```

```python
# Example 8.4 Zafarani's book. Assortativity for ordinal attributes
g_ord = nx.Graph()
g_ord.add_node('A',age=18)
g_ord.add_node('B',age=20)
g_ord.add_node('C',age=21)
g_ord.add_edges_from([('A','C'),('B','C')])
posizione = nx.spring_layout(g_ord)
nx.draw_networkx(g_ord,pos=posizione)
```

```python
print(nx.numeric_assortativity_coefficient(g_ord,'age'))
```

```python
g = nx.Graph()
g.add_nodes_from([1,2,3,4,5,6,7,8,9,])
g.add_edges_from([(1,4),(1,5),(1,6),(1,7),(1,8),(2,9),(2,10),(2,11),(2,12),(3,13),(3,14),(3,15),(3,16),(3,17)])
g.add_edges_from([(1,13),(2,4)])
posizione = nx.spring_layout(g)
nx.draw_networkx(g,pos=posizione)
print(nx.degree_assortativity_coefficient(g))
```
